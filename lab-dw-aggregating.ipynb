```python
def clean_columns(df):
    df.columns = (
        df.columns
        .str.strip()                                        
        .str.replace(r"(?<!^)(?=[A-Z])", "_", regex=True)   
        .str.replace(r"[^\w]+", "_", regex=True)             
        .str.lower()                                         
        .str.replace(r"_+", "_", regex=True)                 
        .str.strip("_")                                      
    )
    return df
df = clean_columns(df)    
df.columns
```

```python
df = df.drop(columns="unnamed_0")
df.columns
```

```python
df.head(1)
```

```python
#Create a new DataFrame that only includes customers who:
#have a low total_claim_amount (e.g., below $1,000),
#have a response "Yes" to the last marketing campaign.
```

```python
low_claim = df["total_claim_amount"] < 1000
responded_yes = df["response"] == "Yes"

df_filtered = df[low_claim & responded_yes]

df_filtered.head(5)
```

```python
#Using the original Dataframe, analyze:
#the average monthly_premium and/or customer lifetime value by policy_type and gender for customers who responded "Yes", and
#compare these insights to total_claim_amount patterns, and discuss which segments appear most profitable or low-risk for the company.
```

```python
analysis = (
    df[df["response"] == "Yes"]          
    .groupby(["policy_type", "gender"])  
    .agg(
        avg_monthly_premium=("monthly_premium_auto", "mean"),
        avg_clv=("customer_lifetime_value", "mean"),
        avg_total_claim_amount=("total_claim_amount", "mean")
    )
    .round(2)  
)

analysis
```

Females of the policy Personal Auto have the highest avg_monthly_premium and the highest avg_customer_lifetime_value
But, it's the 3rd in avg_total_claim_amount, which is bad, but not so far from the average. 
All things considered it represents the most profitable or low-risk for the company.

```python
#The marketing team wants to analyze the number of policies sold by state and month. 
#Present the data in a table where the months are arranged as columns and the states are arranged as rows.
```

```python
df["effective_to_date"].dtype
```

```python
df["effective_to_date"] = pd.to_datetime(df["effective_to_date"], format="%m/%d/%y")
df["month"] = df["effective_to_date"].dt.month_name()
```

```python
df["month"]
```

```python
policies_table = df.pivot_table(
    index="state",
    columns="month",
    values="number_of_policies",
    aggfunc="sum"
)

policies_table
```

```python
#Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.
#Hint:

#To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. 
#Afterwards, you will need to sort the data by the count of policies sold in descending order.
#Next, you will select the top 3 states with the highest number of policies sold.
#Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.
```

```python
total_policies_by_state = df.groupby("state")["number_of_policies"].sum().sort_values(ascending=False)
top_3_states = total_policies_by_state.head(3).index
top_states_df = df[df["state"].isin(top_3_states)]
policies_top3 = top_states_df.pivot_table(
    index="state",
    columns="month",
    values="number_of_policies",
    aggfunc="sum"
)
month_order = ["January", "February"]
existing_months = [m for m in month_order if m in policies_top3.columns]
policies_top3 = policies_top3.reindex(columns=existing_months)
policies_top3 = policies_top3.reindex(index=top_3_states)
policies_top3
```
